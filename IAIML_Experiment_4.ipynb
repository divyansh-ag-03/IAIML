#Pearson Correlation

```python
# Pearson's correlation coefficients
correlation_matrix = np.abs(df1.corr())

# Set a correlation threshold
correlation_threshold = 0.5

# Features with correlation above the threshold to 'subscribers'
selected_features = correlation_matrix['subscribers'][correlation_matrix['subscribers'] > correlation_threshold]

# Filter the dataset to keep only the selected features
X_filtered = df1.loc[:, selected_features.index]  # Use loc to select columns by name

# Print the selected features
print("Selected Features:")
print(selected_features.index)
```

```python
X_filtered
```

**The selected features based on a Pearson's correlation coefficient threshold of 0.5 with respect to the 'subscribers' column are:**

1. **'rank'**: There is a relatively strong positive linear relationship between 'rank' and 'subscribers,' meaning that as the 'rank' increases, the 'subscribers' tend to increase as well.

2. **'subscribers'**: This is the target variable, and its correlation with itself is always 1.

3. **'video views'**: There is a relatively strong positive linear relationship between 'video views' and 'subscribers,' suggesting that as the number of 'video views' increases, the 'subscribers' tend to increase as well.
